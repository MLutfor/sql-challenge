```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
departments_csv = Path("csv_files/departments.csv")
departments_read = pd.read_csv(departments_csv)

dept_emp_csv = Path("csv_files/dept_emp.csv")
dept_emp_read = pd.read_csv(dept_emp_csv)

dept_manager_csv = Path("csv_files/dept_manager.csv")
dept_manager_read = pd.read_csv(dept_manager_csv)

employees_csv = Path("csv_files/employees.csv")
employees_read = pd.read_csv(employees_csv)

salaries_csv = Path("csv_files/salaries.csv")
salaries_read = pd.read_csv(salaries_csv)

titles_csv = Path("csv_files/titles.csv")
titles_read = pd.read_csv(titles_csv)
```
